```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def detect(pic, n=1):
    img = cv.imread(f'opencv/Resources/Photos/{pic}.jpg')
    RGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    haar_cascade = cv.CascadeClassifier('opencv/haar_face.xml')

    faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=n)
    
    plt.imshow(RGB)
    plt.show()
    
    for (x,y,w,h) in faces_rect:
        cv.rectangle(RGB, (x, y), (x+w, y+h), (0,255,0), thickness=2)

    plt.imshow(RGB)
    plt.show()
    
    print(f'{len(faces_rect)} faces detected.')
```

```python
detect('group 1')
```

```python
detect('group 2', 7)
```

```python
detect('group 3', 2)
```
